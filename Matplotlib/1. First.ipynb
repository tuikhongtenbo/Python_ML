```python
x = np.arange(0, 10)
```

```python
x
```

```python
y = x*2
```

```python
y
```

```python
import matplotlib.pyplot as plt 
z = x**4
fig = plt.figure()
plt.plot(x, y)
plt.xlabel('x AXIS')
plt.ylabel('y AXIS')
#LARGE AXES
axes1 = fig.add_axes([0, 0, 1, 1]) 
axes1.plot(x, z)
axes1.set_xlim(0, 5)
axes1.set_ylim(0,  1000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of COAL')
#SMALL AXES
axes2 = fig.add_axes([0.1, 0.3, 0.5, 0.5])
axes2.plot(x, y)
axes2.set_xlim(0, 5)
axes2.set_ylim(0,  20)
axes2.set_xlabel('Ax')
axes2.set_ylabel('Bx')
axes2.set_title('Expenditure of COAL')
plt.show()
```

```python
fig = plt.figure(figsize = (3, 1), dpi = 1000) #size of the figure: dpi = ?; 10 (inch) is the width & 3 (inch) is the height
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, y)
axes1.set_xlim(0, 5)
axes1.set_ylim(0, 10)
axes1.set_xlabel('Ax')
axes1.set_ylabel('By')
fig.savefig('new1figure.png', bbox_inches = 'tight')
```
