```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([6.7, 4.6, 3.5, 5.5])
y=np.array([8.1, 5.6, 4.3, 6.7])
```

```python
x1, y1=3.5, 4.3
x2, y2=6.7, 8.1
#y1 = ax1 + b
#y2 = ax2 + b
# =======> a = (y1 - y2)/(x1 - x2)
# =======> b = y1 - ax1
a=(y1-y2)/(x1-x2)
b=y1 - a*x1
x_values=np.array([6, 4.6, 5.5])
y_values=a * x_values + b
x_line=np.linspace(min(x1, x2), max(x1, x2), 10000)
y_line=a*x_line + b
plt.figure(figsize=(8, 5), dpi=200)
plt.plot()
plt.plot(x_line, y_line, marker='o', markersize=6, label='f(x)')
plt.scatter([x1, x2], [y1, y2], color='red', s=80,label='Original points')
plt.scatter(x_values, y_values, color='yellow', s=300, label='Interpolation points')
for i in range(len(x_values)):
    plt.text(x_values[i], y_values[i], f'({x_values[i]}, {y_values[i]:.2f})', verticalalignment='top') # use .2f instead of round( , )
    plt.plot([x_values[i], x_values[i]], [0, y_values[i]], 'g--')
    plt.plot([0, x_values[i]], [y_values[i], y_values[i]], 'g--')
plt.xlim(3, 7)
plt.ylim(4, 10)
plt.xlabel(r'Area (x 100 m$^2$)')
plt.ylabel(r'Price (Tael)')
plt.legend()
plt.grid(True)
plt.show()
```
