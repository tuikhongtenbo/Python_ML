```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11) 
b = a**4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y)
axes[1].plot(a, b)
```

```python
fig1, axes1 = plt.subplots(nrows = 2, ncols = 2)

axes1[0][0].plot(x, y)
axes1[0][1].plot(y, x)
axes1[1][0].plot(a, b)
axes1[1][1].plot(b, a)
```

```python
fig2, axes2 = plt.subplots(nrows = 3, ncols = 1)

for ax in axes2:
    ax.plot(x, y)
```

```python
fig1, axes1 = plt.subplots(nrows = 2, ncols = 2)

axes1[0][0].plot(x, y)
axes1[0][1].plot(y, x)
axes1[1][0].plot(a, b)
axes1[1][1].plot(b, a)
```

```python
fig1, axes1 = plt.subplots(nrows = 2, ncols = 2, figsize = (4, 10), dpi = 300)

axes1[0][0].plot(x, y)

#Like the plot has studied before
axes1[0][1].plot(y, x)
axes1[0][1].set_title('Hehe')
axes1[0][1].set_xlabel('x hihi')
axes1[0][1].set_xlim(2, 5)

axes1[1][0].plot(a, b)
axes1[1][1].plot(b, a)

fig1.suptitle("MaIn", fontsize = 15) # fontsize is the function that manage the size of the main title
fig1.subplots_adjust(wspace = 0.5, hspace = 0.5)

fig1.set_figwidth(10) #this function manage the width of subplots
fig1.set_figheight(5) #this function manage the height of subplots
plt.tight_layout()
fig1.savefig('SecondSubs.png',  bbox_inches = 'tight')
```
