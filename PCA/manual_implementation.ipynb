```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler_X = scaler.fit_transform(df)
# scaler_X - scaler_X.mean(axis=0)
```

```python
scaler_X
```

```python
covariance_matrix = np.cov(scaler_X, rowvar=False)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
```

```python
eigenvalues.shape
```

```python
eigenvectors.shape
```

```python
num_components = 2
```

```python
np.argsort([2, 1, 3])
```

```python
sorted_key = np.argsort(eigenvalues)[::-1][:num_components]
```

```python
eigenvalues, eigenvectors = eigenvalues[sorted_key], eigenvectors[:, sorted_key]
```

```python
# Original data --> project --> eigenvectors
principal_components = scaler_X @ eigenvectors
```

```python
principal_components
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(principal_components[:, 0], principal_components[:, 1])
```
