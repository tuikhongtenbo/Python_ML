```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\ply58\\PythonML\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\StudentsPerformance.csv')
```

```python
df
```

```python
sns.boxplot(data=df, x='test preparation course', y='math score', hue='test preparation course')
```

```python
plt.figure(figsize=(10, 5), dpi=200)
plt.ylim=(20, 100)
sns.boxplot(data=df, x='parental level of education', y='reading score', 
            hue='parental level of education', palette='Set2')
```

```python
plt.figure(figsize=(10, 10), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education', 
            hue='lunch', palette='bright', inner='stick', split=True)
plt.legend(bbox_to_anchor=(1.24, 0.5))
```

```python
plt.figure(figsize=(8, 6), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=3, hue='test preparation course')
plt.legend(bbox_to_anchor=(1.25, 0.5))
```

```python
plt.figure(figsize=(8, 6), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=3, hue='test preparation course', dodge=True)
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
sns.boxenplot(data=df, x='math score', y='test preparation course', hue='test preparation course', palette='Set2')
```
