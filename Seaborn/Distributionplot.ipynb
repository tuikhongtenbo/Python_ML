```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\ply58\\PythonML\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\dm_office_sales.csv')
```

```python
df
```

```python
plt.figure(figsize=(5, 5), dpi=100)
sns.rugplot(x='salary', data=df, height=0.25)
```

```python
plt.figure(figsize=(7, 5), dpi=100)
sns.set(style='darkgrid')
sns.histplot(data=df, x='salary', bins=50, color='purple', ec='yellow', lw=10, ls='--')
```

```python
plt.figure(figsize=(7, 5), dpi=100)
sns.set(style='darkgrid')
sns.displot(data=df, x='sales', bins=30, color='green', ec='red', lw=20, ls='--', kde=True, rug=True)
```

```python
sns.kdeplot(data=df, x='sales')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 2000)
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sns.set(style='ticks')
sns.displot(data=sample_ages, x='age', rug=True, kde=True, bins=60)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[20, 50], bw_adjust=0.1, shade=True, color='orange')
```
