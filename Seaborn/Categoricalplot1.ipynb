```python
plt.figure(figsize=(10, 5), dpi=100)
sns.countplot(data=df, x='division', hue='division')
```

```python
plt.figure(figsize=(10, 5), dpi=100)
sns.countplot(data=df, x='level of education', hue='level of education', palette='bright')
```

```python
plt.figure(figsize=(10,5), dpi=100)
sns.countplot(data=df, x='level of education', hue='division', palette='Accent')
```

```python
plt.figure(figsize=(10, 5), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', 
            estimator=np.mean, ci='sd', hue='division',
            palette='Set1')
plt.legend(bbox_to_anchor=(1.05, 1))
```
