```python
%matplotlib  inline
from math import erf, factorial
import torch
from IPython import display
from d2l import torch as d2l
```

```python
n = 10
p = 0.2

def binomial(n, k):
    comb = 1
    for i in range(min(k, n - k)):
        comb = comb * (n - i) // (i + 1)
    return comb

pmf = torch.tensor([p**i * (1-p)**(n - i) * binomial(n, i) for i in range(n + 1)])

d2l.plt.stem([i for i in range(n + 1)], pmf, use_line_collection=True)
d2l.plt.xlabel('x')
d2l.plt.ylabel('Probability Mass Function')
```
