```python
p = 0.4

d2l.set_figsize()
d2l.plt.stem([0, 1], [1-p, p], use_line_collection=True)
d2l.plt.xlabel('x')
d2l.plt.ylabel('Probability Mass Function')
d2l.plt.show()
```

```python
x = d2l.arange(-1, 2, 0.01)
def F(y):
    return 0 if y < 0 else 1 if y > 1 else 1 - p
d2l.plot(x, torch.tensor([F(y) for y in x]), 'x', 'Cummulative Density Function')
```
