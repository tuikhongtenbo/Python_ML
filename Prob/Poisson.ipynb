```python
%matplotlib  inline
from math import erf, factorial
import torch
from IPython import display
from d2l import torch as d2l
```

```python
lamda = 5

xs = [i for i in range(20)]
pmf = torch.tensor([torch.exp(torch.tensor(-lamda)) * lamda**k / factorial(k) for k in xs])

d2l.plt.stem(xs, pmf, use_line_collection=True)
d2l.plt.xlabel('x')
d2l.plt.ylabel('Probability Mass Function')
```
