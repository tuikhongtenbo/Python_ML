```python
import numpy as np
import math
import matplotlib.pyplot as plt


x = np.linspace(-5, 5, 100000)
K = ['rectangular', 'triangular', 'parabolic', 'biweigh', 'gaussian', 'silverman']

def I(u):
    return np.where((u >= -1) & (u <= 1), 1, 0)

def visualize(I):
    plt.figure(figsize=(8,6), dpi=150)
    plt.plot(x, I, color='r')

Ix = I(x)
```

```python
print(Ix)
```

```python
def choose(I, K):
    for k in K:
        if k == 'rectangular':
            f = I * 0.5
            visualize(f)

        if k == 'triangular': 
            f = (1 - abs(x)) * I
            visualize(f)
        
        if k == 'parabolic':
            f = 0.75 * (1 - x**2) * I
            visualize(f)
        
        if k == 'biweigh':
            f = 15/16 * ((1 - x**2)**2) * I
            visualize(f)
        
        if k == 'gaussian':
            a = 1 / math.sqrt(2*math.pi)
            f = a * math.exp(-0.5 * x**2)
            visualize(f)
        
        if k == 'silverman':
            a = abs(x) / math.sqrt(2) + math.pi / 4
            b = -abs(x) / math.sqrt(2)
            f = np.dot((0.5 * math.exp(b) * math.sin(a)) * I)
            visualize(f)

choose(Ix, K)
```
