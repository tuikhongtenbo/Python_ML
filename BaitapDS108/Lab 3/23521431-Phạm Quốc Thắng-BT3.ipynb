#### Changing features names

```python
feature_names = X_mean_std.columns
feature_names 
```

```python
# 't' to 'Time domain:'
# 'f' to 'Frequency domain:'
new_features1 = []
for name in feature_names:
    if name[0] == 't':
        new = 'Time domain: ' + name[1:]
    elif name[0] == 'f':
        new = 'Frequency domain: ' + name[1:]
    else:
        new = name
    new_features1.append(new)
feature_names = new_features1
```

```python
# 'mean\\(\\)'  to	'mean value'
# 'std\\(\\)'	 to   'standard deviation value'
# '-X'   to	 'on X axis'
# '-Y'   to	 'on Y axis'
# '-Z'  to	 'on Z axis'
# 'AccJerk'  to	 'acceleration jerk'
# 'Acc'  to	'acceleration'
# 'GyroJerk'  to	'angular velocity jerk'
# 'Gyro'  to	'angular velocity'
# 'Mag'   to	'magnitude'
dict_names = {
    'mean()': 'mean value',
    'std()': 'standard deviation value',
    '-X': ' on X axis',
    '-Y': ' on Y axis',
    '-Z': ' on Z axis',
    'AccJerk': 'acceleration jerk',
    'Acc': 'acceleration',
    'GyroJerk': 'angular velocity jerk',
    'Gyro': 'angular velocity',
    'Mag': 'magnitude'
}

for key, value in dict_names.items():
    feature_names = [name.replace(key, value) for name in feature_names]
# '-' to ','
feature_names = [name.replace('()', ' ').strip() for name in feature_names]
feature_names = [name.replace('-', ', ').strip() for name in feature_names]
```

```python
feature_names = [name.replace(',', ', ') for name in feature_names]
feature_names = [name.replace(':', ': ') for name in feature_names]
```

```python
X_mean_std.columns = feature_names
```
